```python
def route(start_point,end_point,mode="walking") :  ### mode “driving”, “walking”, “bicycling” or “transit”

    gmaps = googlemaps.Client(key='')

    # Geocoding an address
    geocode_result_start = gmaps.geocode(start_point)
    geocode_result_end = gmaps.geocode(end_point)
    # Look up an address with reverse geocoding
    #reverse_geocode_result = gmaps.reverse_geocode((25.1323777197085,121.7438090697085))

    # Request directions via public transit
    now = datetime.now()
    directions_result = gmaps.directions(start_point,
                                         end_point,
                                         mode=mode,
                                         optimize_waypoints =True,)

    
    lat = []
    lon = []
    for i in range(len(directions_result[0]['legs'][0]['steps'])) :

        a = directions_result[0]['legs'][0]['steps'][i]


        try :
            for j in range(len(a['steps'])) :

                b = a['steps'][j]

                lat.append(b['start_location']['lat'])
                lon.append(b['start_location']['lng'])            

                lat.append(b['end_location']['lat'])
                lon.append(b['end_location']['lng'])

                lat.append(b['end_location']['lat'])
                lon.append(b['end_location']['lng'])
        except KeyError :


            lat.append(a['start_location']['lat'])
            lon.append(a['start_location']['lng'])

            lat.append(a['end_location']['lat'])
            lon.append(a['end_location']['lng'])

            lat.append(a['end_location']['lat'])
            lon.append(a['end_location']['lng'])
            pass

    
    df_gas = pd.read_csv('../data/df_gas.csv')
    df = pd.read_csv('../data/real_final.csv')
    #df['recommand'] = (df['推薦保險_1'].replace('None','') + df['推薦保險_2'].replace('None','')).replace('','None')
    df_caf = pd.read_csv('../data/caf_final.csv')
    # creat map
    ### lat lon to tuple
    temp = []
    for i,j in zip(df['lat'],df['lon']) :

        temp.append((i,j))

    df['lat_lon'] = temp
    
    ### filter region by 行政區
     
    route_map = folium.Map(location=[df['lat'].mean(), df['lon'].mean()],
                      zoom_start=16)
    tooltip =utf2asc('點擊顯示詳細信息')
    tooltip_danger =utf2asc('顯勿設施可能影響範圍')
    
    latlon = []
    for i,j in zip(lat,lon) :
        latlon.append([i,j])
    
    
    temp = pd.DataFrame()
    for latlonrotue in latlon:

        temp = pd.concat([temp,df[(abs(df['lat'] - latlonrotue[0]) <= 0.0008) & (abs(df['lon'] - latlonrotue[1]) <= 0.0008)]])

    temp = temp.drop_duplicates()    
    df = temp
    
    df = df.reset_index()
    ### mark all company
    for i in range(len(df)) :
        
        address = df.loc[i,'lat_lon']
        company_name = df.loc[i,'公司名稱']
        cate = df.loc[i,'產業別']
        recommand = df.loc[i,'recommand']
        
        
        
        if '建築物清潔服務業' in cate : 
            
            folium.Marker(location=address,
                      popup=folium.Popup('<p style="font-size:110%;font-family:Microsoft JhengHei;">'+utf2asc(company_name)+'</p>'+
                                         '<p style="color:blue;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(cate)+'</b></p>'+
                                         '<p style="font-size:110%;font-family:Microsoft JhengHei;"><b>'+utf2asc('保險推薦')+'</b></p>'+
                                         '<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(recommand)+'</b></p>',
                                          max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-leaf',color='lightgray'),
                      tooltip=tooltip).add_to(route_map)            
            
            
        
        if '電器承裝業' in cate :
            
            folium.Marker(location=address,
                      popup=folium.Popup('<p style="font-size:110%;font-family:Microsoft JhengHei;">'+utf2asc(company_name)+'</p>'+
                                         '<p style="color:blue;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(cate)+'</b></p>'+
                                         '<p style="font-size:110%;font-family:Microsoft JhengHei;"><b>'+utf2asc('保險推薦')+'</b></p>'+
                                         '<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(recommand)+'</b></p>',
                                          max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-flash',color='darkpurple'),
                      tooltip=tooltip).add_to(route_map)            
            
            
            
            
        if '食品' in cate :
        
            folium.Marker(location=address,
                      popup=folium.Popup('<p style="font-size:110%;font-family:Microsoft JhengHei;">'+utf2asc(company_name)+'</p>'+
                                         '<p style="color:blue;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(cate)+'</b></p>'+
                                         '<p style="font-size:110%;font-family:Microsoft JhengHei;"><b>'+utf2asc('保險推薦')+'</b></p>'+
                                         '<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(recommand)+'</b></p>',
                                          max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-shopping-cart',color='darkred'),
                      tooltip=tooltip).add_to(route_map)        
        
        
        
        if '燃料導管安裝工程業' in cate :
        
            folium.Marker(location=address,
                      popup=folium.Popup('<p style="font-size:110%;font-family:Microsoft JhengHei;">'+utf2asc(company_name)+'</p>'+
                                         '<p style="color:blue;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(cate)+'</b></p>'+
                                         '<p style="font-size:110%;font-family:Microsoft JhengHei;"><b>'+utf2asc('保險推薦')+'</b></p>'+
                                         '<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(recommand)+'</b></p>',
                                          max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-filter',color='black'),
                      tooltip=tooltip).add_to(route_map)        
        

        if '配管工程業' in cate :

            folium.Marker(location=address,
                      popup=folium.Popup('<p style="font-size:110%;font-family:Microsoft JhengHei;">'+utf2asc(company_name)+'</p>'+
                                         '<p style="color:blue;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(cate)+'</b></p>'+
                                         '<p style="font-size:110%;font-family:Microsoft JhengHei;"><b>'+utf2asc('保險推薦')+'</b></p>'+
                                         '<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(recommand)+'</b></p>',
                                          max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-cog',color='darkblue'),
                      tooltip=tooltip).add_to(route_map)

        if '五金' in cate :
            
            folium.Marker(location=address,
                      popup=folium.Popup('<p style="font-size:110%;font-family:Microsoft JhengHei;">'+utf2asc(company_name)+'</p>'+
                                         '<p style="color:blue;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(cate)+'</b></p>'+                                         
                                         '<p style="font-size:110%;font-family:Microsoft JhengHei;"><b>'+utf2asc('保險推薦')+'</b></p>'+
                                         '<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(recommand)+'</b></p>',
                                          max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-wrench',color='gray'),
                      tooltip=tooltip).add_to(route_map)

        if '室內裝潢業' in cate :
            
            folium.Marker(location=address,
                      popup=folium.Popup('<p style="font-size:110%;font-family:Microsoft JhengHei;">'+utf2asc(company_name)+'</p>'+
                                         '<p style="color:blue;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(cate)+'</b></p>'+ 
                                         '<p style="font-size:110%;font-family:Microsoft JhengHei;"><b>'+utf2asc('保險推薦')+'</b></p>'+
                                         '<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(recommand)+'</b></p>',
                                          max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-home',color='lightblue'),
                      tooltip=tooltip).add_to(route_map)

        if '服飾' in cate :
            
            folium.Marker(location=address,
                      popup=folium.Popup('<p style="font-size:110%;font-family:Microsoft JhengHei;">'+utf2asc(company_name)+'</p>'+
                                         '<p style="color:blue;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(cate)+'</b></p>'+
                                         '<p style="font-size:110%;font-family:Microsoft JhengHei;"><b>'+utf2asc('保險推薦')+'</b></p>'+
                                         '<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(recommand)+'</b></p>',
                                          max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-shopping-cart',color='beige'),
                      tooltip=tooltip).add_to(route_map)

    ### mark the gas station nearby
       
    temp = pd.DataFrame()
    for latlonrotue in latlon:

        temp = pd.concat([temp,df_gas[(abs(df_gas['lat'] - latlonrotue[0]) <= 0.001) & (abs(df_gas['lon'] - latlonrotue[1]) <= 0.001)]])

    temp = temp.drop_duplicates()    
    df_gas = temp    
    
    df_gas = df_gas.reset_index()
    
    
    ##### 注意如果有空集合出現的話
    
    for i in range(len(df_gas)) :
    
        folium.Marker(
                    location=[df_gas.loc[i,'lat'], df_gas.loc[i,'lon']],
                    popup=folium.Popup('<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(df_gas.loc[i,'name'])+'</b></p>',
                                          max_width=120),
                    icon=folium.Icon(icon='glyphicon glyphicon-fire',color='red'),
                    ).add_to(route_map)
        folium.Circle(
                    radius=100,
                    location=[df_gas.loc[i,'lat'], df_gas.loc[i,'lon']],
                    popup=folium.Popup('<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(df_gas.loc[i,'name'])+'</b></p>',
                                          max_width=120),
                    tooltip=tooltip_danger,
                    color= 'red',
                    fill = True,
                    ).add_to(route_map)   
    
    
    for i in range(len(df_caf)) :
        
        folium.Marker(location=[df_caf.loc[i,'lat'],df_caf.loc[i,'lon']],
                      popup=folium.Popup('<p style="color:green;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc('國泰產險')+'</b></p>'+
                                         '<p style="color:green;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(df_caf.loc[i,'branch'])+'</b></p>',
                         max_width=120),
                      icon=folium.Icon(icon='glyphicon glyphicon-tree-deciduous',color='green'),
                      ).add_to(route_map)


        folium.Circle(
                    radius=5000,
                    location=[df_caf.loc[i,'lat'],df_caf.loc[i,'lon']],
                    popup=folium.Popup('<p style="color:green;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc('國泰產險')+'</b></p>'+
                                        '<p style="color:green;font-size:115%;font-family:Microsoft JhengHei;"><b>'+utf2asc(df_caf.loc[i,'branch'])+'</b></p>',
                        max_width=120),
                    color='green',fill=False).add_to(route_map)
    
    ### 路徑

    folium.PolyLine(    #polyline方法為將座標用線段形式連線起來
        locations=latlon,    #將座標點連線起來
        weight=8,  #線的大小為3
        color='blue',  #線的顏色為橙色
        opacity=0.8    #線的透明度
    ).add_to(route_map)    #將這條線新增到剛才的區域m內

    ### marker

    ### start point
    
    folium.Marker(location=[geocode_result_start[0]['geometry']['location']['lat'],geocode_result_start[0]['geometry']['location']['lng']],
          popup=folium.Popup('<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+'Start'+'</b></p>'),
          icon=folium.Icon(icon='glyphicon glyphicon-play',color='red'),
          ).add_to(route_map)   
    
    
    ### end point
    
    folium.Marker(location=[geocode_result_end[0]['geometry']['location']['lat'],geocode_result_end[0]['geometry']['location']['lng']],
          popup=folium.Popup('<p style="color:#ff6347;font-size:115%;font-family:Microsoft JhengHei;"><b>'+'End'+'</b></p>',),
          icon=folium.Icon(icon='glyphicon glyphicon-stop',color='red'),
          ).add_to(route_map)       
    
    return route_map
```
