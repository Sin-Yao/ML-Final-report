```python
import urllib.request as Req
import bs4
```

```python
def getdata(url):
    request = Req.Request(url, headers = {
        "User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.113 Safari/537.36"
    })
    with Req.urlopen(request) as response:
        data=response.read().decode("UTF-8")
    root = bs4.BeautifulSoup(data,"html.parser")
    titles = root.find_all("li" ,attrs={"class": "p0 BbHr"})
    for i in titles:
        print(i.text.split('地址:')[0])
        print(i.text.split('|')[1])
        print(i.text.split('|')[2])
        #print(i.text.split('|')[3])
    nextpage = root.find("a",title="下一頁")
    return nextpage["href"]

pageurl ="https://aibee.com.tw/address/%E6%96%B0%E5%8C%97%E5%B8%82/"
c=1
try:
    while c<51:
        pageurl="https://aibee.com.tw"+getdata(pageurl)
        print("page_number: {}".format(c))
        c+=1
except:
    print("Done")
```

```python
def getlink(url):
    request = Req.Request(url, headers = {
        "User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.113 Safari/537.36"
    })
    with Req.urlopen(request) as response:
        data=response.read().decode("UTF-8")
    root = bs4.BeautifulSoup(data,"html.parser")
    nextpage = root.find_all("a",itemprop='url')
    link = []
    for i in nextpage:
         link.append("https://aibee.com.tw"+i["href"])
    return link
def getmenu(url):
    request = Req.Request(url, headers = {
        "User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.113 Safari/537.36"
    })
    with Req.urlopen(request) as response:
        data=response.read().decode("UTF-8")
    root = bs4.BeautifulSoup(data,"html.parser")
    nextpage = root.find("a",title="下一頁")
    return nextpage["href"]

pageurl ="https://aibee.com.tw/address/%E6%96%B0%E5%8C%97%E5%B8%82/"
menu = ["https://aibee.com.tw/address/%E6%96%B0%E5%8C%97%E5%B8%82/"]
c=1
try:
    while c<51:
        menu.append("https://aibee.com.tw" + getmenu(pageurl))
        pageurl="https://aibee.com.tw" + getmenu(pageurl)
        c+=1
except:
    print("Done")
```

```python
detaillink = []
for i in range(1,len(menu)+1):
    #print(i)
    temp = getlink(menu[i-1])
    for j in range(1,len(temp)+1):
            detaillink.append(temp[j-1])
```

```python
def getname(url):
    request = Req.Request(url, headers = {
        "User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.113 Safari/537.36"
    })
    with Req.urlopen(request) as response:
        data = response.read().decode("UTF-8")
    root = bs4.BeautifulSoup(data,"html.parser")
    titles = root.find_all("p",limit=18)
    for i in titles:
        print(i.text)
```
