```python
def getlink(url):
    t = 0.2
    time.sleep(t)
    request = Req.Request(url, headers = {
        "User-Agent":random.choice(my_headers)
    })
    with Req.urlopen(request) as response:
        data=response.read().decode("UTF-8")
    root = bs4.BeautifulSoup(data,"html.parser")
    nextpage = root.find_all("a",itemprop='url')
    link = []
    for i in nextpage:
         link.append("https://aibee.com.tw"+i["href"])
    return link
def getmenu(url):
    t = 0.2
    time.sleep(t)
    request = Req.Request(url, headers = {
        "User-Agent":random.choice(my_headers)
    })
    with Req.urlopen(request) as response:
        data=response.read().decode("UTF-8")
    root = bs4.BeautifulSoup(data,"html.parser")
    nextpage = root.find("a",title="下一頁")
    return nextpage["href"]

pageurl ="https://aibee.com.tw/address/%E6%96%B0%E5%8C%97%E5%B8%82/"
menu = ["https://aibee.com.tw/address/%E6%96%B0%E5%8C%97%E5%B8%82/"]
c=1
try:
    while c<51:
        menu.append("https://aibee.com.tw" + getmenu(pageurl))
        pageurl="https://aibee.com.tw" + getmenu(pageurl)
        c+=1
except:
    print("Done")
```

```python
detaillink = []
for i in range(1,len(menu)+1):
    #print(i)
    temp = getlink(menu[i-1])
    for j in range(1,len(temp)+1):
            detaillink.append(temp[j-1])
```

```python
def getdata(url):
    t = 0.2
    time.sleep(t)
    request = Req.Request(url, headers = {
        "User-Agent":random.choice(my_headers)
    })
    with Req.urlopen(request) as response:
        data = response.read().decode("UTF-8")
    root = bs4.BeautifulSoup(data,"html.parser")
    result = root.find_all("strong",limit=18)
    information=[]
    for i in result:
        information.append(i.text)
    return information
```
